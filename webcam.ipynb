```python
import numpy as np
import cv2
from keras.models import load_model
```

```python
def predict(mod_path):
    model = load_model(mod_path)
    cap = cv2.VideoCapture(0)
    while True:
        ret, frame = cap.read()#changing rgb to gray
        gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        new_img = cv2.resize(gray_img,(28,28))#normalizing data
        new_img = new_img/255
        final_img = np.reshape(new_img,[1,28,28])
        prediction = model.predict(final_img)
        final_ans =np.argmax(prediction)
        cv2.putText(frame, f"Predicted Digit: {final_ans}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow('Webcam', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
predict('pradeep.h5')
```
